```python
#Experiment 3 对原始时间序列进行累加和变换，以及计算各阶累加和变换序列的分形维数
#导入实验需要的包
import datetime
import pandas as pd 
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
import math
#读取文件，提取其中的收盘价序列
#打开一个之前存好数据的文件
fsplit = open('./microsoft_half.csv','r')
#按行读取文件内容到数组
lines_label = fsplit.readlines()
fsplit.close()
#记录股票收盘价的数组
close = list()
#从数组中的每一条记录里提取收盘价的数值，存在close数组中
for each in lines_label[1:]:
    tokens = each.split(',')
    close.append(tokens[4])
#获取的原始数据是时间倒序的，需要调整为按照时间顺序
closed = close[::-1]
#声明代表天数的数组和收盘价数组，构成一段时间序列
#定义代表时间的数组，按照第一天，第二天的顺序[1,2,...]
dateX = np.arange(1,len(closed)+1)
#将收盘价存入numpy数组
X = np.array(closed,dtype='float32')
```

```python
#累加函数:用于累加和变换，公式16
def cirSum(X):
    #用来存储累加结果的数组
    result = list()
    for i in range(0,len(X)):
        #在累加和序列的某一个位置，将原数列上这个位置之前的元素都加起来，称为累加和变换
        result.append(sum(X[0:i+1]))
    return result
```

```python
#对原序列作累加和变换，得到一阶累加和序列
S1 = cirSum(X)
```

```python
#对一阶累加和序列作累加和变换，得到二阶累加和序列
S2 = cirSum(S1)
S2
```

```python
#对二阶累加和序列作累加和变换，得到三阶累加和序列
S3 = cirSum(S2)
S3
```

```python
#计算分形维数D，依据公式13，计算一个D只需要两个点（ri  , Ni）, (rj , Nj)
def calConsDem(X):
    #声明一个数组用于存储利用相邻两点计算得到的D
    consdem = list()
    #从第二天开始，依次选择相邻两天的数据计算D ，第一天不能用来计算，因为没有第0天
    for i in range(2,len(X)):
        #公式13的内容，在双对数坐标系中，求通过两点的直线的斜率
        consdem.append(math.log(X[i]/X[i-1])/math.log(float(i-1)/float(i)))
    return consdem
```

```python
#返回的结果，见论文中Table 3, 各阶累加和序列的分形维数D 
D0 = calConsDem(X)
D1 = calConsDem(S1)
D2 = calConsDem(S2)
D3 = calConsDem(S3)
```
